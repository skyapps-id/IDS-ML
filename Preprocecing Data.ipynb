### Dataset Preprocecing

```python
df['category'] = df['category']+'-'+df['subcategory']
```

```python
a = df.loc[df['category'] == 'DoS-UDP']
b = df.loc[df['category'] == 'DDoS-TCP']
c = df.loc[df['category'] == 'DDoS-UDP']
d = df.loc[df['category'] == 'DoS-TCP']
e = df.loc[df['category'] == 'DDoS-HTTP']
f = df.loc[df['category'] == 'DoS-HTTP']
g = df.loc[df['category'] == 'Normal-Normal']
```

```python
newData = pd.concat([a,b,c,d,e,f,g], axis=0, join='outer', ignore_index=False)
```

```python
newData['category'] = newData['category'].replace(['DoS-TCP'], 'DoS-TCP')
newData['category'] = newData['category'].replace(['DoS-UDP'], 'DoS-UDP')
newData['category'] = newData['category'].replace(['DDoS-TCP'], 'DDoS-TCP')
newData['category'] = newData['category'].replace(['DDoS-UDP'], 'DDoS-UDP')
newData['category'] = newData['category'].replace(['DDoS-HTTP'], 'DDoS-HTTP')
newData['category'] = newData['category'].replace(['DoS-HTTP'], 'DoS-HTTP')
# newData['category'] = newData['category'].replace(['Reconnaissance-Service_Scan-Service_Scan'], 'Reconnaissance-Service-Scan')
newData['category'] = newData['category'].replace(['Normal-Normal'], 'Normal')
```

```python
newData.drop([
  'pkSeqID',
  'stime',
  'flgs',
  'proto',
  'saddr',
  'sport',
  'daddr',
  'dport',
  'state',
  'seq',
  'flgs_number',
  'ltime',
  'attack',
  'stddev',
  'sum',
  'mean',
  'min',
  'max',
  'proto_number',
  'state_number',
  'TnBPSrcIP',
  'TnBPDstIP',
  'TnP_PSrcIP',
  'TnP_PDstIP',
  'TnP_PerProto',
  'TnP_Per_Dport',
  'AR_P_Proto_P_SrcIP',
  'AR_P_Proto_P_DstIP',
  'N_IN_Conn_P_DstIP',
  'N_IN_Conn_P_SrcIP',
  'AR_P_Proto_P_Sport',
  'AR_P_Proto_P_Dport',
  'Pkts_P_State_P_Protocol_P_DestIP',
  'Pkts_P_State_P_Protocol_P_SrcIP',
  'subcategory',
], axis=1, inplace=True)
```
